```python
import pandas as pd
import glob
import os


files = os.path.join("data/", "cgo_20_2021*.csv")

files = glob.glob(files)

data=[]
time=[]
delta=[]

def get_data(name):
    data.append(name)
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
for i in range(0, 4987, 1):
    elem=str(df["Діоксид азоту"][i])
    for j in range(len(elem)-1):
        if elem[j]==',': 
            list1 = list(elem)
            list1[j] = '.'
            elem = ''.join(list1)
    if elem=='nan': data.append('')
    else: data.append(float(elem))
    pos=str(df["Дата"][i])
    hour=int(df["Строк відбору"][i])
    pos=list(pos)
    day=int(pos[8]+pos[9])
    month=int(pos[5]+pos[6])
    restime=hour+day*24+month*24*30
    time.append(restime)
df
```

```python
import math

delta_time=[]

def getdelta():
    for i in range(len(data)-1):
        if(data[i] != '' and data[i + 1] != '' and i%4==1):
            delta.append(round(float(data[i]) - float(data[i + 1]), 6))
            delta_time.append(time[i])
                
getdelta()

delta_time
```

```python
import matplotlib.pyplot as plt
plt.scatter(delta_time, delta, marker='.')
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

model = LinearRegression()
y = pd.DataFrame(delta)
x = pd.DataFrame(delta_time)
model.fit(x, y)
plt.scatter(x, y, marker='.')
plt.plot(x, model.predict(x), color='orange') 
```
